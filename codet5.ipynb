```python
from transformers import T5ForConditionalGeneration, T5Tokenizer

# Pastikan model dan tokenizer dimuat dengan benar
model_name = "Yaziddd/indo-t5-small-medical-qa"
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)

def generate_qa_pairs(context):
    # Prompt untuk QA Generation
    prompt = f"Buatlah satu pertanyaan dan jawabannya berdasarkan konteks berikut:\n\n{context}"
    
    # Tokenisasi input
    inputs = tokenizer.encode(prompt, return_tensors="pt", max_length=512, truncation=True)
    
    # Generate output
    outputs = model.generate(inputs, max_length=128, num_beams=4, early_stopping=True)
    
    # Decode output menjadi teks
    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    
    # Pisahkan antara Question dan Answer
    if "Q:" in generated_text and "A:" in generated_text:
        try:
            question, answer = generated_text.split("A:")
            question = question.replace("Q:", "").strip()
            answer = answer.strip()
            return {"question": question, "answer": answer}
        except ValueError:
            # Jika format tidak sesuai
            return {"question": None, "answer": generated_text}
    else:
        return {"question": None, "answer": generated_text}
```

```python
import pandas as pd

# Membaca dataset
df = pd.read_csv('D:/TUGAS AKHIR/TASI-2425-113/cleaned_data/cleaned_data.csv')

# Hasilkan QA pairs untuk setiap context
results = []
for index, context in enumerate(df['context']):
    try:
        qa_result = generate_qa_pairs(context)
        results.append({
            "context": context,
            "question": qa_result.get("question"),
            "answer": qa_result.get("answer")
        })

        # Tampilkan hasil pada konsol
        print(f"Context ke-{index + 1}:")
        print(f"Context: {context[:100]}...")
        print(f"Question: {qa_result.get('question')}")
        print(f"Answer: {qa_result.get('answer')}\n")
    except Exception as e:
        print(f"Error pada context ke-{index + 1}: {context[:50]}... -> {e}")
        results.append({
            "context": context,
            "question": None,
            "answer": None
        })

# Membuat DataFrame dari hasil
results_df = pd.DataFrame(results)

# Menampilkan hasil dalam bentuk tabel
print(results_df)
```
